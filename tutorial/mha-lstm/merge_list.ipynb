```python
class ListNode:
    def __init__(self,value=0,next=None):
        self.value = value
        self.next = next

def merge_two_sort_lists(l1,l2):
    dummy = ListNode(0)
    current = dummy

    #merge the listnode
    while l1 and l2:
        if l1.value < l2.value:
            current.next = l1
            l1 = l1.next
        else: 
            current.next = l2
            l2 = l2.next
        current = current.next

    if l1:
        current.next = l1
    if l2:
        current.next = l2

    return dummy.next

def print_list(head):
    current = head
    while current:
        print(current.value, end="->")
        current = current.next
    print("END")

# l1 = ListNode(1)
# l1.next = ListNode(3)
# l1.next.next = ListNode(5)

# l2 = ListNode(2)
# l2.next = ListNode(4)
# l2.next.next = ListNode(6)

# merge_list = merge_two_sort_lists(l1,l2)

# print("Merge and sort the list")
# print_list(merge_list)



# 1:all empty:
# print("Test 1: Both empyt")
# print_list(merge_list(None,None))

# merge_list = merge_two_sort_lists(l1,l2)

# print("Merge and sort the list")
# print_list(merge_list)


# 2:Only one empty:
print("Test 1: Both empyt")
l1 = None

merge_list = merge_two_sort_lists(l1,l2)

print("Merge and sort the list")
print_list(merge_list)


# # 3:Not same len:
# print("Test 1: Both empyt")

# l1 = ListNode(1)
# l1.next = ListNode(3)
# l1.next.next = ListNode(5)



# l2 = ListNode(2)
# l2.next = ListNode(4)
# l2.next.next = ListNode(6)
# l2.next.next.next = ListNode(8)



# merge_list = merge_two_sort_lists(l1,l2)

# print("Merge and sort the list")
# print_list(merge_list)

# # 3:Have the same len but some overlapping


# l1 = ListNode(1)
# l1.next = ListNode(3)
# l1.next.next = ListNode(5)



# l2 = ListNode(1)
# l2.next = ListNode(2)
# l2.next.next = ListNode(5)




# merge_list = merge_two_sort_lists(l1,l2)


# # 4: One is in other

# l1 = ListNode(1)
# l1.next = ListNode(3)
# l1.next.next = ListNode(5)



# l2 = ListNode(1)
# l2.next = ListNode(3)
# l2.next.next = ListNode(5)
# l2.next.next.next = ListNode(7)




# merge_list = merge_two_sort_lists(l1,l2)
```
